```python
arr = [12, 35, 1, 10, 34, 1]

first = second = -1

for num in arr:
    if num > first:
        second = first
        first = num
    elif num != first and num > second:
        second = num

print(second)
```

```python
# Q3.Given an array of integers, rotating array of elements by k elements either left or right.
def reverse(arr, l, r):
    while l < r:
        arr[l], arr[r] = arr[r], arr[l]
        l += 1
        r -= 1


arr = [1, 2, 3, 4, 5]
k = 2
n = len(arr)

k = k % n

reverse(arr, 0, k-1)
reverse(arr, k, n-1)
reverse(arr, 0, n-1)

print(arr)
```
