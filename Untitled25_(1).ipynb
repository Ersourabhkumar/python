# String in Python

```python
# String is immutable datatype . Immutable means we cannot change or update values at a runtime.
# Indexing start from 0 and  length start from 1 .
# forward Indexing start from 0 and reverse/backward indexing start from -1.
```

```python
a = 'sam'
print(a)
type(a)
```

```python
# Indexing
a = 'I love Maths'
print(a[0])      ####forward Indexing start from 0
print(a[1])
print(a[2])
print(a[3])
print(a[4])
print(a[5])
```

```python
print(a[-1])   ###Reverse Indexing
print(a[-2])
print(a[-3])
print(a[-4])
print(a[-5])
```

# Slicing in String `

str[start_index_value:last_index_value:step_value]

start_index_value  will include and last_index_value will exclude .

```python
a = 'I love Maths'
print(a[2:6])
print(a[7:12])
print(a[7:])
print(a[:7])
print(a[:])
print(a[0:11:2]) ###[start:last:step]
print(a[0:11])
```

# String Functions

# (1). lower() and upper(): Convert the string to lowercase or uppercase.


```python
text = "Hello"
print(text.lower())
print(text.upper())
```

```python
a = 'Maths'
a.upper()
```

```python
# (2). strip(): Removes whitespace from both ends of a string.
text = "   Hello, World!   "
print(text.strip())
```

```python
a = 'Maths  '  ##rstrip() will remove whitespaces from the right side .
a.rstrip()
```

```python
a = ' Maths'   ###lstrip() will remove whitespaces from left side .
a.lstrip()
```

```python
# (3). replace(): Replaces a substring with another string.
text = "I love Python"
print(text.replace("Python", "coding"))
```

```python
# (4). split(): Splits a string into a list.
text = "apple, banana, cherry"
print(text.split(", "))
```

```python
# (5). join(): Joins elements of a list into a single string.
fruits = ['apple', 'banana', 'cherry']
print(", ".join(fruits))
```

```python
# (6). Isalnum() : It will return True if we have alphabats , numbers , both , otherwise return False.
a = "sam123"
print(a.isalnum())
```

```python
# (7). isdigit() : It will return True when we have only digits in a string .
a = '134'
print(a.isdigit())
```

```python
# (8). isalpha() : It will return True when we have only alphabats in a string .
a = 'sam'
print(a.isalpha())
```
