```python
#nested_loop
for i in range(1,10):
    print("python student",i)
    for  j in range(1,5):
        print("assignment",j)
```

```python
for i in range(10,0,-1):
    print("python student",i)
    for  j in range(1,9):
        print("student",i ,"assignment",j)
```

```python
count=1
for i in range(10,0,-1):
    print("python student",i)
    for  j in range(1,9):
        print("student",i ,"assignment",j, count)
        count+=1
```
