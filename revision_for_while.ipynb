```python
# Reverse a Number
n=12345
rev=0
```

```python
while n>0:
  s=n%10
  rev=rev*10+s
  n//=10
print(rev)
```

```python
# Palindrome Number
n=131
temp=n
t=0
while n>0:
  s=n%10
  t=t*10+s
  n//=10
if n == t:
  print("palidrome")
```
